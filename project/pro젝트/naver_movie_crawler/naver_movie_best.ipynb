```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
from pprint import pprint
    
    #영화 코드 알아내는 코드
    
search_url = 'https://search.naver.com/search.naver'
query = input('영화 제목: ')
params = {
    'where': 'nexearch',
    'query':query
}
search_resp = requests.get(search_url,params = params)
movie_code = ''
search_soup = bs(search_resp.content)

i = 99
try:
    search_movie_url = search_soup.find('div',class_='info_main').find('h3').a['href']
    for idx,val in enumerate(search_movie_url):
        if val == '=':
            i = idx
        if idx > i:
            movie_code += val
    print(movie_code)

    movie_url = 'https://movie.naver.com/movie/bi/movie/bi/mi/script.nhn'
    code=movie_code
    params = {
        'code' : code
    }
    movie_resp = requests.get(movie_url,params=params)
    movie_soup = bs(movie_resp.content)
    ul_content = movie_soup.find('ul',class_='lines').contents
    new_list = []
    new_ul=[]

    for ul in ul_content:
        if ul == "\n":
            continue
        new_ul.append(ul)


    for diag in new_ul:
        new_dict = {}
    #     print(diag.find('p',class_='char_part'))
        new_dict['명대사'] = diag.find('p',class_='one_line').text
        if diag.find('p',class_='char_part') == None:
            new_dict['배우'] = ''
        else:
            new_dict['배우']= diag.find('p',class_='char_part').find('a').text
        if diag.find('p',class_='char_part') == None:
            new_dict['배역'] = ''
        else:
            new_dict['배역'] = diag.find('p',class_='char_part').find('span').text

        new_list.append(new_dict)
    pprint(new_list)
    best_diag_file = pd.DataFrame(new_list)
    best_diag_file.to_csv(f'{query}명대사파일.csv',encoding = 'cp949')
except Exception as e:
    print('영화제목을 제대로 쳐주세요')

    #영화 명대사 알아내는 코드
```
