```python
import requests
from bs4 import BeautifulSoup
```

```python
import urllib
```

```python
movie_url = "https://movie.naver.com/movie/bi/mi/detail.nhn"
params = dict(
    where='post',
    code="136900"
)
```

```python
resp = requests.get(movie_url,params=params)
resp
```

```python
soup = BeautifulSoup(resp.content, 'html.parser')
soup
```

```python
#제목추출
get_title1 = soup.find('div', class_='mv_info_area')
get_title2 =get_title1.find('div', class_='mv_info')
get_title3 =get_title2.find('h3', class_='h_movie')
Movie_title = get_title3.find('a').text
```

```python
post_content = soup.find('div', class_="made_people")
post_content
```

```python
#배우리스트
name_list = post_content.find_all('a', class_= "k_name")
name_list
actor_list= []
for i in name_list:
    actor_list.append(i.text)
actor_list    


```

```python
#역할 리스트
roll_list = post_content.find_all('p', class_= "pe_cmt")

roll = []
for j in roll_list:
    roll.append(j.text)
roll
new_roll = [] 
for k in range(len(roll)):
    new_roll.append(str(roll[k].strip('\n')))
    
    
    
```

```python
end_game= []
for i in range(len(new_roll)):
    end_game.append(actor_list[i] + ': ' + new_roll[i] )
     
Movie_title    
end_game
```

```python
print('제목: ', Movie_title)
print('배우,역할: ', end_game)
```
