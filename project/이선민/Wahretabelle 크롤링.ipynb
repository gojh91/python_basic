## Wahretabelle 크롤링

0708 - 1819 오프사이드 기록 가져오기

```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint
from selenium import webdriver
from pprint import pprint
import pandas as pd


# 크롬창 열기
driver_path = 'C:\\SEONMIN\\chromedriver_win32\\chromedriver'
driver = webdriver.Chrome(driver_path)
url = 'https://www.wahretabelle.de/statistik/abseits'


year_change = ['?saisonId=1','?saisonId=51','?saisonId=80','?saisonId=83','?saisonId=86','?saisonId=89','?saisonId=93','?saisonId=202','?saisonId=205','?saisonId=302','?saisonId=307',' ']
saison = ['0708','0809','0910','1011','1112','1213','1314','1415','1516','1617','1718','1819']

for idx, year in enumerate(year_change):
    
    driver.get(url+year)


    # 자료 가져오기
    offside_even = driver.find_elements_by_class_name('even')
    offside_odd = driver.find_elements_by_class_name('odd')

    temp_list = []
    sum_list=[]


    count = 0
    even_num = 0
    odd_num=0
    temp_list = []

    for i in range(18):        
        if count % 2 == 1:

            temp_list.append(offside_even[even_num].text)
            count += 1
            even_num += 1
#             print(count)


        elif count % 2 == 0:

            temp_list.append(offside_odd[odd_num].text)
            count += 1
            odd_num += 1
#             print(count)


    print(temp_list)       

    v1 = driver.find_element_by_id('yw0_c4').text
    v2 = driver.find_element_by_id('yw0_c5').text
    
    temp_rank = []
    temp_name = []
    v1_list =[]
    v2_list = []
    
    
    
    for i in temp_list:
        split = i.split(' ')
        split.remove('')
        split.remove('')

        temp_rank.append(split[0])
        temp_name.append(split[1:3])
        v1_list.append(int(split[-2]))
        v2_list.append(int(split[-1]))

    temp_dict = {
        'name': temp_name, 
        saison[idx]+v1: v1_list, 
        saison[idx]+v2 : v2_list
    }    

    temp_df = pd.DataFrame(temp_dict, index = temp_rank)
    sum1 = temp_df[saison[idx]+v1].sum()
    sum2 = temp_df[saison[idx]+v2].sum()
    
    sum_df_index = ['total']
    sum_df = pd.DataFrame({'name':'sum',saison[idx]+v1 : sum1, saison[idx]+v2 : sum2},index = sum_df_index)
    temp_df = temp_df.append(sum_df, ignore_index=False)
    print(temp_df) 
    
    temp_df.to_csv('offside_'+saison[idx]+'.csv')
```
