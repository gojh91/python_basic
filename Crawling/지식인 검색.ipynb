# URL 이동 + 지식인 크롤링 

```python
query = "호국보훈의 달"

BASE_URL = 'https://search.naver.com/search.naver'
params = {
    'where': 'kin',
    'query': query,
    'kin_start': 1
}
```

```python
kin_list = []

for i in range(1, 50, 10):
    params['kin_start'] = i
    
    resp = requests.get(BASE_URL, params)
    soup = BeautifulSoup(resp.content)
    ul_tag = soup.find('ul', class_='type01')
    li_tag_list = ul_tag.find_all('li')
    
    for li_tag in li_tag_list:
        anchor = li_tag.find('dt').find('a')

        title = anchor.text
        link = anchor['href']
    
        content_box = li_tag.find_all('dd')
    
        qs = content_box[1].text
        answer = content_box[2].text
    
        kin_dict = {
            'title': title,
            'link':link,
            'qs': qs,
            'answer':answer
        }
        kin_list.append(kin_dict)
    
```

# Function 으로 이쁘게 만들기

```python
def parse_kin_page(url, params):
    kin_list = []
    
    resp = requests.get(url, params)
    soup = BeautifulSoup(resp.content)
    ul_tag = soup.find('ul', class_='type01')
    li_tag_list = ul_tag.find_all('li')
    
    for li_tag in li_tag_list:
        anchor = li_tag.find('dt').find('a')

        title = anchor.text
        link = anchor['href']
    
        content_box = li_tag.find_all('dd')
    
        qs = content_box[1].text
        answer = content_box[2].text
    
        kin_dict = {
            'title': title,
            'link':link,
            'qs': qs,
            'answer':answer
        }
        kin_list.append(kin_dict)
    return kin_list
```

```python
kin_list = []

for i in range(1, 50, 10):
    params['kin_start'] = i
    kin_list.extend(parse_kin_page(BASE_URL, params))
```

```python
def crawl_kin_with_range(url=BASE_URL, params=PARAMS, start_page=1, end_page=30):
    kin_list = []
    
    for i in range(start_page, end_page, 10):
        params['kin_start'] = i
        kin_list.extend(parse_kin_page(BASE_URL, params))
    return kin_list
```

```python
kin_list = crawl_kin_with_range(params=params, start_page=1, end_page=30)
```
